## Notebook 1: Data Cleaning and Preprocessing

This notebook performs data cleaning and preprocessing for all datasets used in the NorthStar case study.

The objectives are to:
- Load all raw CSV files directly from GitHub.
- Assess data quality issues.

- Handle missing values and duplicates.

### Import Libraries

```python
# Import required libraries
import pandas as pd
import numpy as np
from pathlib import Path

# Libraries for saving and downloading files
import os
import shutil
from google.colab import files

# Display all columns when viewing DataFrames
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print('Libraries imported successfully.')
```

## Define GitHub Repository Paths

The raw CSV files are loaded directly from the GitHub repository using raw file URLs.

### Set GitHub Base URL

```python
# Replace this with your GitHub username
github_username = "haamnaaa"

# Repository name
repo_name = "DBA-Files"

# Create the base URL for the raw_data folder
base_url = f"https://raw.githubusercontent.com/haamnaaa/DBA-Files/main/raw_data/northstar_dataset/"

print("Base URL:")
print(base_url)
```

## Define the Dataset Files

```python
# List of all dataset files
file_names = [
    "app_events.csv",
    "complaints.csv",
    "customers.csv",
    "deliveries.csv",
    "drivers.csv",
    "hubs.csv",
    "incidents.csv",
    "orders.csv",
    "vehicles.csv"
]
```

## Load All CSV Files

Each dataset is loaded into a dictionary called `data`

```python
import pandas as pd

# Create an empty dictionary to store datasets
data = {}

# Load each CSV file from GitHub
for file_name in file_names:
    dataset_name = file_name.replace(".csv", "")
    url = base_url + file_name

    print("Loading:", url)

    df = pd.read_csv(url)
    data[dataset_name] = df

    print(dataset_name, "loaded successfully.")
    print("Shape:", df.shape)
    print()
```

## Preview the First Five Rows of Each Dataset

```python
# Display first five rows of each dataset
for name in data:
    print("\n" + "=" * 60)
    print("Dataset:", name)
    print("=" * 60)
    display(data[name].head())
```

## Check Data Quality

This section checks:

- Dataset dimensions
- Duplicate rows
- Missing values
- Data types

```python
# Loop through each dataset and display data quality information
for name in data:
    df = data[name]

    print("\n" + "=" * 60)
    print("Data Quality Report:", name)
    print("=" * 60)

    # Shape
    print("Shape:", df.shape)

    # Duplicate rows
    print("Duplicate rows:", df.duplicated().sum())

    # Missing values
    print("\nMissing values in each column:")
    print(df.isnull().sum())

    # Data types
    print("\nData types:")
    print(df.dtypes)
```

## Clean the Data

The following cleaning steps are applied to every dataset:

1. Standardise column names.
2. Remove duplicate rows.
3. Remove extra spaces from text values.
4. Fill missing values.
5. Convert date and time columns to datetime format.

```python
# Create a dictionary to store cleaned datasets
cleaned_data = {}

# Loop through each dataset
for name in data:
    # Make a copy of the original dataframe
    df = data[name].copy()

    # ---------------------------------------------------
    # 1. Standardise column names
    # ---------------------------------------------------
    df.columns = df.columns.str.strip()          # Remove spaces
    df.columns = df.columns.str.lower()          # Convert to lowercase
    df.columns = df.columns.str.replace(" ", "_")  # Replace spaces with underscores

    # ---------------------------------------------------
    # 2. Remove duplicate rows
    # ---------------------------------------------------
    df = df.drop_duplicates()

    # ---------------------------------------------------
    # 3. Clean text columns
    # ---------------------------------------------------
    text_columns = df.select_dtypes(include="object").columns

    for col in text_columns:
        df[col] = df[col].str.strip()

    # ---------------------------------------------------
    # 4. Fill missing values
    # ---------------------------------------------------
    for col in df.columns:
        # If column is text
        if df[col].dtype == "object":
            df[col] = df[col].fillna("Unknown")
        else:
            # If column is numeric, fill with median
            if df[col].isnull().sum() > 0:
                df[col] = df[col].fillna(df[col].median())

    # ---------------------------------------------------
    # 5. Convert date/time columns
    # ---------------------------------------------------
    for col in df.columns:
        if "date" in col or "time" in col:
            df[col] = pd.to_datetime(df[col], errors="coerce")

    # Save cleaned dataframe
    cleaned_data[name] = df

    print(name, "cleaned successfully.")
```

## Verify the Cleaned Data

```python
# Check the cleaned datasets
for name in cleaned_data:
    df = cleaned_data[name]

    print("\n" + "=" * 60)
    print("Cleaned Dataset:", name)
    print("=" * 60)

    print("Shape:", df.shape)
    print("Duplicate rows:", df.duplicated().sum())
    print("Total missing values:", df.isnull().sum().sum())
```

## Save Cleaned CSV Files

The cleaned datasets are saved in a folder named `cleaned_data`.

```python
# Create folder to store cleaned files
os.makedirs("cleaned_data", exist_ok=True)

# Save each cleaned dataframe
for name in cleaned_data:
    file_path = "cleaned_data/" + name + "_clean.csv"
    cleaned_data[name].to_csv(file_path, index=False)

    print("Saved:", file_path)
```

## Create ZIP File and Download

All cleaned CSV files are compressed into a ZIP file and downloaded.

```python
# Create ZIP file
shutil.make_archive("northstar_cleaned_data", "zip", "cleaned_data")

# Download ZIP file
files.download("northstar_cleaned_data.zip")
```
